# Einstein heat capacity 

It was early known that the heat capacity for the most solids can be described by:
$C = 3 k_B$. This is known as the Dulong-Petit law.

Einstein wondered in 1907 why this law doesn't hold for low temperatures.
So he assumed that every atom is in a harmonic well created by the interactions with his neighbors. He also assumed that every atom is in a identical harmonic well and all have the frequency $\omega$. This frequency is also known as the "Einstein" frequency.

#### Derivation of the Einstein heat capacity

It is well known that energy eigenstates of a single harmonic oscillator is given by:
$E_n = \hbar \omega (n+1/2)$.

Starting from this the partition can be calculated:
$Z_{1D} = \sum_{n = 0}^{\infty}e^{-\beta \hbar \omega(n+1/2)} = \frac{1}{2\sinh(\beta\hbar\omega/2)}$

The next step to get the heat capacity is to calculate the expectation of energy. This is done by the derivative of the logarithm of the partition function in respect to $\beta$. The result is: $<E> = \hbar \omega(n_B(\beta\hbar\omega)+\frac{1}{2}$, where $n_B(\beta\hbar\omega)$ is the Bose-Einstein statistic.

#### Final result for the Einstein heat capacity

Next step is to take the derivative of this expaction of energy with respect to T. The final heat capacity C is:
$C=k_B(\beta\hbar\omega)^2\frac{e^{\beta\hbar\omega}}{(e^{\beta\hbar\omega}-1)^2}$.

#### Start to explore the features of this result

First let us plot this heat capacity to check if for high T this C can recover the Dulong-Petit law.

```python
import math 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# define the function of the heat capacity
# set kB=1=hbar, beta=1/T
def heat_cap(omega, T):
    return np.square(omega/T)*np.exp(omega/T)/np.square((np.exp(omega/T) - 1))
```

```python
#define parameters
dt=0.01
T = np.arange(0.1,300,dt)
omega = 20
```

```python
plt.figure(figsize=(10,10))
plt.plot(T/omega,heat_cap(omega,T),label='Einstein heat capacity')
plt.axhline(y=1,color='k', linestyle='--',label='law of Dulong-Petit')
plt.xlabel('T/omega',Fontsize=20)
plt.ylabel('C',Fontsize=20)
plt.xlim([0,10])
plt.title('Einstein heat capacity')
plt.grid(True)
plt.legend(loc=5,prop={'size': 25})
plt.show()
```

```python
# as you can see the Einstein heat capacity goes for high T into the Dulong-Petit law
```

```python
#as you can see the C vanishes rapidly as T->0. This can be understood as the degrees of freedom "freeze out"
```
