# 1D chain with two masses

![Title](graz_chain.png)

Source and more information: [TU Graz](http://lampx.tugraz.at/~hadley/ss1/phonons/1d/1d2m.php)

The idea is to briefly sketch the solution for this problem and then plot the dispersion relation in the first Brilloun zone to explore the dependance of the dispersion relation on these parameters.

We want to consider the simple 1D chain with two basis atoms with different mass $M_1$ and $M_2$. All atoms are equally spaced by $\frac{a}{2}$. Since the coupling between the atoms is supposed to be harmonic the EOM can be derived with Newton's law:

$M_1 \frac{d^2 u_l}{dt^2} = C( v_{l-1}-2u_l+v_l)$

$M_2 \frac{d^2 v_l}{dt^2} = C( u_{l}-2v_l+u_{l+1})$

$M_1$ is the mass of the heavy atom and $M_2$ the mass of the light atom. $u_l$ the displacement of the heavy atom and $v_l$ the displacement of the light atom. C is just the spring constant. To find the normal modes as usual the Ansatz:

$u_l = u_k e^{i(kla-\omega t)}$

$v_l = v_k e^{i(kla-\omega t)}$

is done. Plugging this Ansatz back into the EOM leads to the matrix equation:

$\left[ \begin{array}{rrrr}
\omega^2M_1-2C & C(1+e^{-ika})  \\
C(1+e^{+ika}) & \omega^2M_2-2C  \\
\end{array}\right]  \left[ \begin{array}{rrrr}
v_k  \\
u_k   \\
\end{array}\right] = 0 $.



This leads to the dispersion relation:

$\omega^2 = C(\frac{1}{M_1}+\frac{1}{M_2})\pm C\sqrt{(\frac{1}{M_1}+\frac{1}{M_2})^2-\frac{4\sin(ka/2)^2}{M_1M_2}} $

### Plotting

```python
#import
import math 
from math import pi as pi
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#define parameters
#define ration between M_1 and M_2
ratio = 1 #define ratio M2/M1 here!
M1=1
M2 = ratio*M1
C=2
k = np.arange(-pi, pi, 0.02)
def omega_optical(M1,M2,C,k):
    return np.sqrt(C*(1/M1+1/M2)+C*np.sqrt(np.square(1/M1+1/M2)-4*np.square(np.sin(k/2))/(M1*M2)))/np.sqrt(2*C*(1/M1+1/M2))
def omega_acc(M1,M2,C,k):
    return np.sqrt(C*(1/M1+1/M2)-C*np.sqrt(np.square(1/M1+1/M2)-4*np.square(np.sin(k/2))/(M1*M2)))/np.sqrt(2*C*(1/M1+1/M2))
      
```

```python
plt.figure(figsize=(10,8))
plt.plot(k,omega_acc(M1,M2,C,k),label='Optical branch') 
plt.plot(k,omega_optical(M1,M2,C,k),label='Accustic branch')
plt.xlabel('k*a')
plt.ylabel('omega/(omega_max)')
plt.title('Dispersion relation for the mass ratio of '+str(ratio))
plt.grid(True)
plt.legend(prop={'size': 20})
plt.show()
```
