## Density of states

This notebook visualizes the dependence of the density of single-particle states of energy once for phonons in three dimensions and once for a free electron gas in three dimensions.

### Phonons
The density of orbitals for phonons in three dimensions is given by:  
$D(\epsilon)= \frac{V}{2 \pi^2}Â  (\frac{2m}{\hbar^2})^{3/2}  \epsilon^{1/2}$.
To get the actual density of filled states we have to multiplicate the Bose-Einstein statistic ($f(\epsilon)=\frac{1}{\exp(\epsilon/(k_B T))-1}$).
**[Why to use the Bose-Einstein statistic?]**

```python
# import the necessary libraries
import math 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
#Define the function piecewise
def bose_einstein(E, T): #Define the Bose-Einstein statistic
    return 1/(np.exp(E /T) - 1)

def rho(E):    #define the density of states, set all parameters beside E to 1
    rho = np.power(E,1/2);
    return rho;

def bose_rho(E, T): # define the occupation of the density of states
    return np.multiply(rho(E),bose_einstein(E, T));
```

```python
#define parameters, all energy scales in scales of kb
dt=0.01
E = np.arange(0.02,5,dt)
T=1
```

## Free electron gas
The DoS is given by the same expression: $D(\epsilon)= \frac{V}{2 \pi^2}Â  (\frac{2m}{\hbar^2})^{3/2}  \epsilon^{1/2}$, but now we have to use the Fermi-Dirac statistic ($f(\epsilon)=\frac{1}{\exp([(\epsilon-\mu)/(k_B T))]+1)}$).

```python
#Define the function piecewise
def fermi_dirac(E, Î¼, T): #define the Fermi-Dirac statistic
    return 1/(np.exp((E - Î¼)/T) + 1)

def rho(E):    #define the density of states, set all parameters beside E to 1
    rho = np.power(E,1/2);
    return rho;

def fermi_rho(E, Î¼, T):   #define the occupation of the density of states
    return fermi_dirac(E, Î¼, T)*rho(E);
```

```python
#define parameters, all energy scales in scales of kb
dt=0.01
Î¼ = 5
T=0.2
E = np.arange(0,Î¼*1.5,dt)
```

```python
#plot the Density of states and the actual occupation
plt.figure(figsize=(20,10))
plt.fill_between(E, fermi_rho(E, Î¼, T),0,color=(0.5, 0.5, 0.5, 0.5))
plt.plot(E, fermi_rho(E, Î¼, T),color='r',label='Occupation number')
plt.plot(E, rho(E),color=(0.5, 0.5, 0.5, 0.5),label='Density of states')
plt.vlines(Î¼, ymin=0, ymax=rho(Î¼), colors='k', linestyles='dashed')
plt.xticks([0,Î¼], [0, 'Î¼'])
plt.xlabel('Energy -->',Fontsize=20)
plt.ylabel('Density of orbitals, relative scale',Fontsize=20)
plt.legend(loc='upper left',fontsize=20)
plt.ylim(0,np.max(rho(E)))
plt.xlim(0,np.max(E))
plt.show()
```
