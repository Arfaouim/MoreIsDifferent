## Density of states in a box

This notebook trys to visualize the results from exercise 6: Density of states in a box.
If you have **not** solved the exercise 6 on your own, please do this first, since the ability of counting modes is crucial in condensed matter physics. In this whole notebook all prefactors are set to one, so one can focus on the various dependencies depending on the dimensionality.
##### Calculating the number of states N in dependance of the quantum number $n$


```python
import math 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def N_1D (n):    #define function for the number of states in 1D, input is an array of quantum number n >=0
    N = n
    return N;
def N_2D (n):    #define function for the number of states in 2D, input is an array of quantum number n >=0
    N = np.square(n)
    return N;
def N_3D (n):    #define function for the number of states in 3D, input is an array of quantum number n >=0
    N = np.power(n,3)
    return N;
```

```python
dn = 0.05 #set the step size for the quantum number n
n = np.arange(0,20,dn) #set the quantum number n in steps dn
fig, axs = plt.subplots(1, 3, figsize=(20, 7.5)) #visualize the functions for the number of states
axs[0].plot(n, N_1D(n)) #1D plot
axs[1].plot(n, N_2D(n)) #2D plot
axs[2].plot(n, N_3D(n)) #3D plot

axs[0].set_xlabel('n',fontsize=20) # set xlabel
axs[1].set_xlabel('n',fontsize=20)
axs[2].set_xlabel('n',fontsize=20)
axs[0].set_ylabel('N(n)',fontsize=20) # set ylabel
axs[1].set_ylabel('N(n)',fontsize=20)
axs[2].set_ylabel('N(n)',fontsize=20)
axs[0].set_yticks([]) # set y ticks
axs[1].set_yticks([])
axs[2].set_yticks([])
axs[0].set_title('1D',fontsize = 20)
axs[1].set_title('2D',fontsize = 20)
axs[2].set_title('3D',fontsize = 20)
fig.tight_layout()
plt.show()
```

#### Next an expression for $N(\epsilon)$ is shown.

```python
def N_1D_eps (eps):    #define function for the number of states  in 1D, input is an array of the energy epsilon eps >=0
    N = np.sqrt(eps)
    return N;
def N_2D_eps (eps):    #define function for the number of states in 2D, input is an array of the energy epsilon eps >=0
    N = eps
    return N;
def N_3D_eps (eps):    #define function for the number of states in 3D, input is an array of the energy epsilon eps >=0
    N = np.power(eps,3/2)
    return N;
```

```python
deps = 0.05 #set the step size for the energy eps
eps = np.arange(0,20,deps) #set the energy in steps of deps
fig, axs = plt.subplots(1, 3, figsize=(20, 7.5)) #visualize the functions for the number of states
axs[0].plot(n, N_1D_eps(eps)) #1D plot
axs[1].plot(n, N_2D_eps(eps)) #2D plot
axs[2].plot(n, N_3D_eps(eps)) #3D plot

axs[0].set_xlabel('n',fontsize=20) # set xlabel
axs[1].set_xlabel('n',fontsize=20)
axs[2].set_xlabel('n',fontsize=20)
axs[0].set_ylabel('N(eps)',fontsize=20) # set ylabel
axs[1].set_ylabel('N(eps)',fontsize=20)
axs[2].set_ylabel('N(eps)',fontsize=20)
axs[0].set_yticks([]) # set y ticks
axs[1].set_yticks([])
axs[2].set_yticks([])
axs[0].set_title('1D',fontsize = 20)
axs[1].set_title('2D',fontsize = 20)
axs[2].set_title('3D',fontsize = 20)
fig.tight_layout()
plt.show()
```

Finally by taking the derivative of the the number of states the density of states is derived. The result is visualized below. 
**What is the general dependance in d-dimensions?**

```python
def rho_1D (eps):    #define function for the number of states  in 1D, input is an array of the energy epsilon eps >=0
    rho = 1/np.sqrt(eps)
    return rho;
def rho_2D (eps):    #define function for the number of states in 2D, input is an array of the energy epsilon eps >=0
    rho = 10*eps/eps;
    return rho;
def rho_3D (eps):    #define function for the number of states in 3D, input is an array of the energy epsilon eps >=0
    rho = np.power(eps,1/2)
    return rho;
```

```python
deps = 0.05 #set the step size for the energy eps
eps = np.arange(0.1,20,deps) #set the energy in steps of deps
fig, axs = plt.subplots(1, 3, figsize=(20, 7.5)) #visualize the functions for the number of states
axs[0].plot(eps, rho_1D(eps)) #1D plot
axs[1].plot(eps, rho_2D(eps)) #2D plot
axs[2].plot(eps, rho_3D(eps)) #3D plot

axs[0].set_xlabel('n',fontsize=20) # set xlabel
axs[1].set_xlabel('n',fontsize=20)
axs[2].set_xlabel('n',fontsize=20)
axs[0].set_ylabel('N(eps)',fontsize=20) # set ylabel
axs[1].set_ylabel('N(eps)',fontsize=20)
axs[2].set_ylabel('N(eps)',fontsize=20)
axs[0].set_yticks([]) # set y ticks
axs[1].set_yticks([])
axs[2].set_yticks([])
axs[0].set_title('1D',fontsize = 20)
axs[1].set_title('2D',fontsize = 20)
axs[2].set_title('3D',fontsize = 20)
fig.tight_layout()
plt.show()
```
