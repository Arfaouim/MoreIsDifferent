# Here we want to discuss the features and characteristics of the Bose-Einstein statistic for Bosonsand Fermi-Dirac statistic for fermions


For high temperatures both statistics go into the classical Boltzmann statistics.

## Bose-Einstein statistics

```python
import math 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#Define the statistics
def bose_einstein(E, μ, T):
    return 1/(np.exp((E - μ)/T) - 1)
```

```python
#define parameters, all energy scales in scales of kb
#notice E > μ
dt=0.01
μ = 10
E = np.arange(μ+0.1,5*μ,dt)
T=20
```

```python
plt.figure(figsize=(10,10))
plt.axvline(x=0,color='k', linestyle='--',label='E=μ')
plt.plot(E-μ,bose_einstein(E,μ,T),label='Bose-Einstein statistic')
plt.xlabel('E-μ')
plt.ylabel('<n(E)>')
plt.title('Bose-Einstein statistics')
plt.grid(True)
plt.legend(prop={'size': 15})
plt.show()
```

## Fermi statistic

```python
#Define the statistics
def fermi_dirac(E, μ, T):
    return 1/(np.exp((E - μ)/T) + 1)
```

```python
#define parameters, all energy scales in scales of kb
dt=0.01
μ = 10
E = np.arange(-T*μ,(T+2)*μ,dt)
T=10
```

```python
#Plot the statistic and explore the width of the "stepfunction"
```

```python
plt.figure(figsize=(10,10))
plt.plot(E-μ,fermi_dirac(E,μ,T),label='Fermi-Dirac statistic')
plt.axvline(x=-2*T,color='k', linestyle='--',label='(E-μ)-2T')
plt.axvline(x=+2*T,color='r', linestyle='--',label='(E-μ)+2T')
plt.xlabel('E-μ')
plt.ylabel('<n(E)>')
plt.title('Fermi-Dirac statistics')
plt.grid(True)
plt.legend(prop={'size': 15})
plt.show()
```

```python
#Plot the statistics for different temperatures T
#You see that as T ->0 the stepfunction at E-μ is received
```

```python
#define parameters, all energy scales in scales of kb
dt=0.01
μ = 10
E = np.arange(-T*μ,(T+2)*μ,dt)
T=10
```

```python
plt.figure(figsize=(10,10))
plt.plot(E-μ,fermi_dirac(E,μ,T),label='T') 
plt.plot(E-μ,fermi_dirac(E,μ,T-5),label='T-5')
plt.plot(E-μ,fermi_dirac(E,μ,T-8),label='T-8')
plt.plot(E-μ,fermi_dirac(E,μ,T-T+0.00001),label='T=0')
plt.xlabel('E-μ')
plt.ylabel('<n(E)>')
plt.title('Fermi-Dirac statistics')
plt.grid(True)
plt.legend(prop={'size': 20})
plt.show()
```
